```python
%pylab inline
import numpy as np
```

```python
num_attributes = 10  # called 'N'
interaction = 3  # called 'K'

payoffs = np.random.rand(num_attributes, 2**interaction)
payoffs
```

```python
def fitness(attributes, interaction, payoffs):
    score = 0
    for i in range(len(attributes)):
        relevant_attributes = attributes[:interaction]
        key = int(''.join([str(c) for c in relevant_attributes]), 2)
        score += payoffs[i, key]
        attributes = np.roll(attributes, 1)  # easy way to handle the end overlap
    return score
```

```python
attributes = np.random.binomial(1, .5, size=num_attributes)
print(attributes)
print(fitness(attributes, interaction, payoffs))
```

c. In NK model, if two actions A1 and A2 overlap in M elements, define the action
distance between the two |A1-A2|=(N-M)/N. Define the expected absolute
reward difference between the two as E(|r(A1)-r(A2)|). Qualitatively graph
expected absolute reward distance as a function of action distance, and how that
depends on K. How does the number of local peaks and the size of basins of
attraction for each peak change with this function?

- if k=0, then for any action distance, the reward distance will be 0.
- if action distance=0, then reward distance is 0, if action distance is N, then 
- if k=n, then the draws are completely random, there is no difference between the expected value of r(a1) and r(a2), so as E(X - Y) = E(X) - E(Y), then the expected reward distance will be 0.


```python
num_attributes = 50
interaction = 3

payoffs = np.random.rand(num_attributes, 2**interaction)

action_distance = []
reward_distance = []
for _ in range(5000):
    a1 = np.random.binomial(1, .5, size=num_attributes)
    a2 = np.random.binomial(1, .5, size=num_attributes)
    M = sum(a1+a2 != 1)
    action_distance.append((num_attributes - M)/num_attributes)
    reward_distance.append(abs(fitness(a1, interaction, payoffs) - fitness(a2, interaction, payoffs)))
    
```

```python
plt.plot(action_distance, reward_distance, '.', alpha=.3)
```

```python
num_attributes = 10
interaction = 10

payoffs = np.random.rand(num_attributes, 2**interaction)

action_distance = []
reward_distance = []
interactions = []
for _ in range(5000):
    interaction = int(np.random.uniform(1,11.1))
    interactions.append(interaction)
    a1 = np.random.binomial(1, .5, size=num_attributes)
    a2 = np.random.binomial(1, .5, size=num_attributes)
    M = sum(a1+a2 != 1)
    action_distance.append((num_attributes - M)/num_attributes)
    reward_distance.append(abs(fitness(a1, interaction, payoffs) - fitness(a2, interaction, payoffs)))
```

```python
plt.plot(interactions, reward_distance, '.', alpha=.2)
```
