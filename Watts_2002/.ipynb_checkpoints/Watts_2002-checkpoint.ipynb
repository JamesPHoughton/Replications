```python
%pylab inline
import networkx as nx
import numpy as np
```

```python
def setup(n_agents=100,
          mean_degree=5,
          active_probability=.03):

    # we construct a network of n agents, in which each agent is connected to k neighbors with
    # probability P_k and the average number of neighbors is <k> = z.
    g = nx.gnp_random_graph(n_agents, mean_degree/n_agents)

    # the population is initially all-off (state 0) and is perturbed at time t = 0 by a small 
    # fraction Phi_0 << 1 of vertices that are switched on 
    nx.set_node_attributes(g, name='state', 
                           values={i:np.random.binomial(1, active_probability) for i in range(n_agents)})
    
    # each agent is assigned a threshold phi drawn at random from a distribution f(phi) that is defined
    # on the unit interval and normalized such that integral from 0 to 1 of f(phi)*dphi = 1, 
    # but which is otherwise arbitrary. 
    nx.set_node_attributes(g, name='threshold', 
                           values={i:.2 for i in range(n_agents)})
    #nx.set_node_attributes(g, 'threshold', {i:np.random.rand() for i in range(n_agents)})
    
    return g

g = setup()
nx.draw_networkx(g, node_color=['r' if g.node[i]['state'] else 'b' for i in g], alpha=.5)
plt.title('Red = Active; Blue = Inactive')
g.node[1]
```

```python
def simulate(g, t_steps=100):
    for t in range(t_steps):
        # all vertices updating their states in random, asynchronous order
        for n in np.random.permutation(g.nodes()):
            if g.node[n]['state'] == 0:
                # An individual agent observes the current states (either 0 or 1) of... its neighbors, 
                # and adopts state 1 if at least a threshold fraction phi of its k neighbors are in state 1,
                # else it adopts state 0.
                neighbor_activation = np.mean([g.node[i]['state'] for i in nx.neighbors(g, n)])
                g.node[n]['state'] = neighbor_activation >= g.node[n]['threshold']
            else:
                # Once a vertex has switched on, it remains on (active) for the duration of the dynamics.
                pass     
    return g
```

```python
plt.figure(figsize=(15,10))
t_steps = 40
for t in range(t_steps):
    plt.subplot(int(np.ceil(t_steps/8)), 8, t+1)
    g = simulate(g, t_steps=1)
    nx.draw_networkx(g, node_color=['r' if g.node[i]['state'] else 'b' for i in g], alpha=.5)
    plt.box('off')
```

```python
%%timeit
simulate(g)
```
